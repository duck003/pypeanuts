Consider the following example

````python
# execute first to create a canvas

from ipyturtle import Turtle
turtle = Turtle(fixed=False, width=500, height=500)
turtle
````

````python
# execute second to draw on canvas
from random import randint
from math import sqrt, cos, pi

class Triangle:
    def __init__(self, side_length, angle):
        self.side = side_length
        self.angle = 180 - (180 - angle)/2
        self.radians = pi*angle/180
        
    def draw(self, t):
        t.forward(self.side)
        t.left(self.angle)
        # this is [the law of cosines](https://en.wikipedia.org/wiki/Law_of_cosines)
        t.forward(sqrt(2*(self.side**2) - 2*(self.side**2)*cos(self.radians)))
        t.left(self.angle)
        t.forward(self.side)
        

TRIS = 30
ANGLE = 360/TRIS
turtle.reset()
for i in range(TRIS):
        tri = Triangle(randint(30, 250), ANGLE)
        tri.draw(turtle)
        turtle.left(180)
````
